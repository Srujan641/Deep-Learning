```python
data = pd.get_dummies(
    df,
    columns=["Gender", "Subscription Type", "Contract Length"],
    drop_first=True
)
```

```python
data.head()
```

```python
data = data.replace({True: 1, False: 0})
```

```python
data.head()
```

```python
data.dtypes
```

```python
col_to_scal=["Tenure","Usage Frequency","Support Calls","Payment Delay","Total Spend","Last Interaction","Age"]
from sklearn.preprocessing import MinMaxScaler
scal=MinMaxScaler()
data[col_to_scal]=scal.fit_transform(data[col_to_scal])
data.head()
```

```python
y=data["Churn"]
x=data.drop(columns="Churn")
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=50)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from tensorflow import keras
import tensorflow as tf 
```

```python
model = keras.Sequential([
    keras.layers.Dense(10,input_shape=(12,),activation="relu"),
    keras.layers.Dense(1,activation="sigmoid")]
)
model.compile(
    optimizer="adam",loss="BinaryCrossentropy",metrics=["accuracy"]
)
with tf.device('/GPU:0'):
    model.fit(x_train,y_train,epochs=5)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pre_ac=model.predict(x_test)
y_pre=[]
for i in y_pre_ac:
    if i>=0.5:
        y_pre.append(1)
    else:
        y_pre.append(0)
y_pre[:5]
```

```python
import seaborn as sb
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_pre)
sb.heatmap(cm,annot=True,cmap="Blues")
```
