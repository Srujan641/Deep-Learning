```python
import tensorflow as tf 
from tensorflow import keras
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv("sonar_dataset.csv",header=None)#if header is non the n columns nmae will be just the columns number
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
df[60]
```

```python
y_bol =pd.get_dummies(df[60],drop_first=True)
y_bol.sample(5)
y = y_bol.iloc[:, 0].astype(int)
y.head()
```

```python
df.drop(columns=60,inplace=True)
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df,y,test_size=0.2)
```

```python
model=keras.Sequential([
    keras.layers.Dense(40,input_dim=60,activation="relu"),
    keras.layers.Dense(20,activation="relu"),
    keras.layers.Dense(1,activation="sigmoid")
])
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
model.fit(x_train,y_train,epochs=100)
```

```python
model.evaluate(x_test,y_test)
```

```python
model_drop=keras.Sequential([
    keras.layers.Dense(40,input_dim=60,activation="relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(20,activation="relu"),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(1,activation="sigmoid")
])
model_drop.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
model_drop.fit(x_train,y_train,epochs=100)
```

```python
model_drop.evaluate(x_test,y_test)
```
